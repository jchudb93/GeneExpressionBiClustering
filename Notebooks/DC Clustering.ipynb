# Clustering

```python
import gc
import pandas as pd
import glob
import sklearn
import numpy as np
```

```python
dcFiles=glob.glob("../Dc/*")
dcFiles
```

```python
trainIndexFiles=glob.glob("../Train/*")
```

```python
dcFiles
```

```python
frames = list()
for f in trainIndexFiles:
    indexes = pd.read_csv(f,sep=";")
    indexList=list(indexes['0'])
    fpath = '../Dc/'+f.split("/")[-1].split(".")[0]+".csv"
    dcRead = pd.read_csv(fpath,sep=";")
    dcPartialTrain=dcRead[dcRead.columns[5:]].loc[indexList].copy()
    del dcRead
    gc.collect()
    frames.append(dcPartialTrain)
```

```python
dcTrain = pd.concat(frames)
```

```python
len(dcTrain)
```

```python
dcTrain.to_csv("../FullTrain/FullTrainDC/FullTrainDC.csv")
```

```python
dcTrain.head()
```

```python
dcTrain.describe()
```

```python
lstn_clusters=range(2,len(dcTrain)+1)
```

```python
import time
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import silhouette_score
```

```python
file = open('../ClusteringOutputs/DC_Clustering/dcMiniBatch.txt','w')  
```

```python
max_silhouette=0
nClusters=2
t0 = time.time()

for n in lstn_clusters:
    mbk=MiniBatchKMeans(init='k-means++', n_clusters=n
                        , max_no_improvement=10, verbose=0)

    mbk.fit(dcTrain)
    labels = mbk.labels_
    s_score = silhouette_score(dcTrain, labels, metric='euclidean')
    
    file.write("n clusters: " + str(n))
    file.write("n silhouette_score" + str(s_score))
    if s_score>max_silhouette:
        max_silhouette = s_score
        nClusters=n
t_mini_batch = time.time() - t0    
print("Num clusters " + str(n))
print("Silhouette score " + str(max_silhouette))
print("Running time " + str(t_mini_batch))
file.close()
    
```
