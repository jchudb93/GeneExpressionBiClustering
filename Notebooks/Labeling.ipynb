```python
df['TargetLabel'] = np.nan
```

```python
##from multiprocessing import Process
i=0
for index,row in df.iterrows():
    if (row['Pvalue']<=0.01):
        df.loc[i,'TargetLabel'] =1
    elif (row['Pvalue']>=0.5):
        df.loc[i,'TargetLabel'] =0
    i+=1
    
```
